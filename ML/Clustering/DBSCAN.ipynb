```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
mms = StandardScaler()
X = mms.fit_transform(X)
```

```python
clusterizer = DBSCAN()
clusterizer.fit(X)
labels = clusterizer.labels_
```

```python
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels, palette=['green', 'red', 'blue']);
```

```python
hdbscan = HDBSCAN()
hdbscan.fit(X)
labels = hdbscan.labels_
```

```python
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels, palette=['green', 'red', 'blue']);
```
