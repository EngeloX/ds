# R2 

Коэффициент детерминации - Статистика **R-квадрат** варьирует в интервале от 0 до 1 и измеряет долю вариации в данных, объясняемую в модели.\
Знаменатель TSS - пропорционален дисперсии отклика($y$)

$$
R^2 = 1 - \frac{RSS}{TSS}
$$
$RSS$ - Остаточная сумма квадратов \
$TSS$ - Общая сумма квадратов

$$
RSS = \sum{(y_{true}-y_{pred})^2}
$$

$$
TSS = \sum(y_{true}-\bar{y_{true}})
$$

$\bar{y_{true}}$ - Среднее значение $y$

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
RSS = np.sum((y_test - y_pred)**2)
RSS
```

```python
TSS = np.sum((y_test - np.mean(y_test))**2)
TSS
```

```python
1-RSS/TSS
```

```python
TSS/np.var(y_test)
```
