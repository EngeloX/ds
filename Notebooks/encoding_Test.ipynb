# LabelEncoder

Преобразует n категорий в числа от 0 до n-1. Принимает только одномерный объект\
Не подходит для категорий с более чем двумя классами и отсутствием внутренней иерархии, в таком случаи столбец просто превратится в числовой количественный столбец.

для порядковых категориальных данных этот способ вряд ли подойдет, потому что LabelEncoder не видит порядка в данных.

````python
df_LE_train = df_train.copy(deep=True)
df_LE_test = df_test.copy(deep=True)
````

````python
df_LE_train['Target']
````

````python
sk_LabelEncoder = sk.LabelEncoder()
sk_LabelEncoder.fit(df_LE_train.loc[:, 'Target'])

df_LE_train['Target'] = sk_LabelEncoder.transform(df_LE_train.iloc[:, -1])
df_LE_train
````
