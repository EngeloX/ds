# Math Statistics

## Типы данных

**Типы данных признаков**
\
**Количественные(Числовые)** - Числовые признаки, могут быть объективно измерены\
**`Непрерывные`** - Любые численные величины, как целые так и дробные(Скорость автомобиля, рост, вес)\
**`Дискретные`** - Только целые величины(Количество покупателей, количество товаров)

\
**Качественные(Категориальные)** - Признаки, представляющие из себя конкретные категории, и могут быть оценены только субьективно\
**`Порядковые`** - Категории между которыми можно установить некую иерархию или упорядочить, могут быть представлены числами(ID пользователя)\
**`Номинативные`** - Категории без иерархии и упорядоченности. (Город, семейное положение)\
**`Бинарные`** - Категории представленные лишь двумя значениями, если не одно то другое(Пол)\
\
**Текстовые** - Признаки, содержащие в себе абзацы текста\
**Дата** - Признаки, выраженные датой и временем



## Оценки центрального положения

**Меры центральной тенденции**

**`Мода (Mode)`** : Наиболее часто встречающееся значение в выборке. \

**`Среднее (mean)`**: Сумма всех значений, деленная на их количество. \
*Синонимы*: среднее арифметическое. 

**`Среднее взвешенное (weighted mean)`**: Сумма произведений всех значений на их веса, деленная на сумму весов. \
*Синонимы*: среднее арифметическое взвешенное. 

**`Среднее усеченное (trimmed mean)`** - Среднее значение, считаемое после отбрасывания фиксированного количества предельных значений (с обоих концов) Например 10 процентов с обоих сторон

**`Медиана(Median)`**: 50%-Квантиль. Середина нечетной выборка, среднее двух центральных элементов в четной выборке. По половине значений находится выше и ниже данного значения.

**`Медиана взвешенная (weighted median)`**: — это такое значение, которое делит упорядоченный набор данных на две части, в каждой из которых сумма весов примерно равна 50% от общей суммы весов.


**`Мат.ожидание `**  :

**`Выброс (robust) `** : Значение сильно отличающееся от большинства данных. Предельное или аномальное значение.\
**`Робастый (outlier) `** : Нечуствительный к выбросам.

```python
import numpy as np
heights = np.array([167, 167, 167, 172, 176, 178, 171, 182, 185, 185])
weights=[2, 2, 2, 1, 1, 2, 1, 2, 3, 3]
```

**Мода**

```python
import scipy.stats
scipy.stats.mode(heights)
```

**Среднее арифметическое**

```python
np.mean(heights)
```

**Среднее арифметическое взвешенное**

```python
np.average(heights, weights=weights)
```

**Среднее усеченное**

```python
from scipy.stats import trim_mean
trim_mean(heights, 0.3)
```

**Медиана**

```python
np.median(heights)
```

**Медиана взвешенная**

```python
import wquantiles
wquantiles.median((heights), weights=weights)
```

## Оценки вариабельности

**`Отклонения (deviations) `** : Разница между наблюдаемыми значениями и оценкой центрального положения. \
*Синонимы*: ошибки, погрешности.

**`Дисперсия (variance) `** : Сумма квадратичных отклонений от среднего, деленная количество.\
Разница между фактическим значением и средним в квадрате, деленная на количество значений\
*Синонимы*: среднеквадратическое отклонение, среднеквадратическая ошибка.

**`Стандартное отклонение (standard deviation)`** : Мера разброса данных относительно их среднего значения. Оно показывает, насколько данные в выборке в среднем отклоняются от среднего значения.\
Корень из дисперсии.\
*Синонимы*: норма l2 , евклидова норма. \

**`Среднее абсолютное отклонение MAD (mean absolute deviation) `**: Среднее абсолютных значений отклонений от среднего \
*Синонимы*: норма l1 , манхэттенская норма.\

**``Медианное абсолютное отклонение от медианы (median absolute deviation from the median) ``** : \
Медиана абсолютных значений отклонений от медианы. 

**`Размах (range) `** : Разница между самым крупным и самым малым значениями в наборе данных.\

**`Квантиль (quantile) `** : Значение делящее отсартированную выборку на части с заданной долей.\
Показывает какое значение заданная доля выборки не превышает.\
**`Квартиль (quartile) `** Квартиль, чья доля задана четвертями "кварталом"(25%, 50%, 75%, 100%)\
**`Перцентиль (percentile) `** : Квантиль задающийся процентом. \
**`Децимиль (decimile) `** :  Квантили, с долей в 10% \
**`Межквартильный размах МКР/IQR (interquartile range) `** : Разница между 75-м и 25-м процентилями. 

**Дисперсия(Вариабильность)**

```python
np.var(heights)
```

**Стандартное отклонение**

```python
np.std(heights)
```

```python
np.sqrt(np.var(heights))
```

**Среднее абсолютное отклонение | MAD**

**Медианное абсолютное отклонение**

**Квантиль, Перцентиль, Квартиль, Децимиль**

```python
# Отсортированная выборка
sorted_heights = np.sort(heights)
sorted_heights
```

```python
# Квантиль 0.8 (80% выборки)
np.quantile(heights, 0.8)
```

```python
# Перцентиль 50% - соответствует медиане
np.percentile(heights, 50), np.median(heights)
```

```python
# Квартили
np.quantile(heights, 0.25), np.quantile(heights, 0.5), np.quantile(heights, 0.75), np.quantile(heights, 1)
```

```python
# Децимиль
np.quantile(heights, 0.1)
```

**Интерквартильный размах | IQR**

```python
Q1 = np.quantile(heights, 0.25)
Q3 = np.quantile(heights, 0.75)
IQR = Q3-Q1
Q1, Q3, IQR
```

### Корреляция

**`Корреляция`** : Линейная взаимосвязь двух или более случайных величин.\
**`Коэффициент корреляции`** : Показатель взаимосвязи. Метрика, которая измеряет степень, с которой числовые переменные связаны друг с другом.\
*Нормирована* в диапазоне от -1 до 1 \

**`Корреляционная матрица`** : Таблица, в которой строки и столбцы — это переменные, а значения ячеек — корреляции между этими переменными.

\
\
**`Диаграмма рассеяния (scatterplot)`**: График, в котором по оси $x$ откладываются значения одной переменной, а по оси $y$ — значения другой. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
x1, y1 = make_regression(n_samples=50, n_features=1, n_informative=1, noise=15, random_state=3)
x2, y2 = make_regression(n_samples=50, n_features=1, n_informative=1, noise=15, random_state=3)
y2 = y2*-1
x3, y3 = make_regression(n_samples=50, n_features=1, n_informative=1, noise=250, random_state=3)

fig, ax = plt.subplots(1, 3, figsize=(12, 4))
ax[0].scatter(x1, y1)
ax[0].grid()
ax[0].set_title('Положительная корреляция')

ax[1].scatter(x2, y2)
ax[1].grid()
ax[1].set_title('Отрицательная корреляция')


ax[2].scatter(x3, y3)
ax[2].grid()
ax[2].set_title('Корреляция отсутствует либо нелинейная');
```

## Распределение и выборки

`Популяция` : Крупный набор данных. \
`Выборка (sample)` : Подмножество из большего по размеру набора данных(из популяции). \
`N(n)` : Размер выборки/популяции. \
`Случайный отбор (random sampling)` : Извлечение данных в sample из популяции в случайном порядке. \
`Стратифицированный отбор (stratified sampling) ` : Разделение популяции на подгруппы(страты) и случайный отбор элементов из каждой страты в выборку. \

`Смещение (bias)` : Систематическая ошибка. \
`Смещенная выборка (sample bias)` : Выборка, которая представляет популяцию в искаженном виде.

----------
`Центральная предельная теорема (central limit theorem)` : Тенденция выборочного распределения принимать нормальную форму по мере увеличения размера данных(выборки). 

-----------

### Нормальное распределение

Нормальное распределение или распределение Гаусса.

`Ошибка (Error)` : Разница между точкой данных и предсказанным либо средним значением. \
`Стандартизировать` : Вычислить среднее значение и разделить  на стандартное отклонение. \
`Z-score (Z-оценка)` : Результат стандартизации отдельной точки (точка-среднее значение/стандартное отклонение) \
`QQ-plot (Квантиль-Квантильный график)` : График позволяющий визуализировать степень близости выборочного распределения к нормальному распределению.

**Нормальное распределение** \


### Длиннохвостые распределения

Бóльшая часть данных не является нормально распределенной. 

`Хвост (tail)` : Длинная узкая часть частотного распределения, где относительно предельные значения встречаются с низкой частотой. \
`Скошенность/Ассиметрия (screw)` : Состояние когда один хвост распределения длиннее другого.

### t-Распределение

Распределение Стьюдента, или t-распределение — это распределение нормальной формы, но немного толще и длиннее в хвостах.
