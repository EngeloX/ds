### Градиентный спуск параболы

```python
import matplotlib.pyplot as plt
import numpy as np
digits = np.arange(-100, 100)


# Функция параболы
def f(x):
    return x**2

# Производная к ней(Градиент функции из одного аргумента)
def gradf(x):
    return 2*x
    
start_point = 10 # Стартовая точка
flag = True # Флаг
eps = 0.0001 # Критерий сходимости

while flag:
    next_point = start_point - gradf(start_point) * 0.1
    print(next_point)
    if abs(start_point - next_point) <=eps:
        flag=False
    start_point = next_point
```

```python
plt.grid()
plt.scatter(next_point, f(next_point), c='r');
plt.plot(digits, f(digits));
```
